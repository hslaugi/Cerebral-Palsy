```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('Results_TD_CP.csv')

metrics = ['dti_fa', 'md', 'ad', 'rd']
groups = ['TD_ND', 'TD_D', 'CSC_LA', 'CSC_MA', 'PV_LA', 'PV_MA']
```

```python
print(f"{'Metrics':<10} | {'p-value':<12} | {'Significant?'}")
print("-" * 40)

for metric in metrics:
    subset = df[df['Metrics Name'] == metric]
    data_to_compare = []
    for group in groups:
        clean_values = subset[group].dropna().values
        data_to_compare.append(clean_values)
    
    try:
        stat, p_val = stats.kruskal(*data_to_compare)
        is_significant = "Yes" if p_val < 0.05 else "No"
        print(f"{metric:<10} | {p_val:.2e}     | {is_significant}")
        
    except ValueError:
        print(f"{metric:<10} | Not enough data to test")
```
